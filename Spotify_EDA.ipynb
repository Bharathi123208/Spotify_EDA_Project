```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
numeric_df = df_tracks.select_dtypes(include=['number'])

# Compute Pearson correlation
corr_df = numeric_df.corr(method="pearson")

# Plot the heatmap
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".2f", vmin=-1, vmax=1, center=0, cmap="inferno", linewidth=1, linecolor="Black")
plt.show()
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(df_tracks))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the plot
plt.figure(figsize=(10, 6))
sns.regplot(data = sample_df, y="loudness", x="energy", color="c").set(title="Loudness vs Energy Correlation")

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y= "popularity", x= "acousticness", color="c").set(title = "Popularity vs Acousticness Correlation")
```

```python
genre = df_tracks[['genre', 'duration']].dropna()
```

```python
plt.title('Duration of songs in different Genres')
sns.color_palette('cividis', as_cmap=False)
sns.barplot(y='genre', x='duration', data = genre)
plt.xlabel('Duration in milliseconds')
plt.ylabel('Genres')
```

```python
# Ensure 'popularity' exists in the DataFrame
if 'popularity' in df_tracks.columns:
    genre = df_tracks[['genre', 'popularity']].dropna()
else:
    print("The 'popularity' column is missing in the DataFrame.")
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
popular = genre.sort_values('popularity', ascending=False).head(10)
sns.barplot(y = 'genre', x = 'popularity', data = popular,hue='genre', palette='viridis', legend=False).set(title='Top 5 Genres by Popularity')
```
